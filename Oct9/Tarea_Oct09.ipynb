Use a mean of 70 and a sigma=2

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import seaborn as sns 
import math 

np.random.seed(201)
```

### Graficas

```python
mu=70
sigma=2
```

```python
x_grafica = np.arange(63, 78, .01)
n1=np.zeros(len(x_grafica))
acumulado=np.zeros(len(x_grafica))
ac=0
for i in range(len(x_grafica)):
   n1[i]=1/(sigma*math.sqrt(2*math.pi))*np.exp(-((x_grafica[i]-mu)*(x_grafica[i]-mu))/(2*sigma*sigma))
   ac=ac+n1[i]
   acumulado[i]=ac
plt.plot(x_grafica,n1)
plt.plot(x_grafica,acumulado/acumulado.max())


```

### Poisson

```python
k = np.arange(0,30, 1)
f=np.zeros((len(lamda),len(k)))
acumulado=np.zeros((len(lamda),len(k)))
ac=0
lamda=[1,4,6,10]
for j in range (len(lamda)):
   for i in range(len(k)):  
      f[j,i]=(np.exp(-lamda[j])*lamda[j]**(k[i]))/(math.factorial(k[i]))
      ac=ac+f[j,i]
      acumulado[j,i]=ac

plt.plot(k,f[0,:], label='F', color='blue')
plt.plot(k,f[1,:], label='F', color='red')
plt.plot(k,f[2,:], label='F', color='yellow')
plt.plot(k,f[3,:], label='F', color='green')

plt.xlabel('k')
plt.ylabel('p(k)')
plt.title('Funcin de distribucin poisson')
plt.legend(('lamda=1', 'lamda=4', 'lamda=6', 'lamda=10'), loc='upper right')
```

```python
plt.plot(k,acumulado[0,:]/acumulado[0,:].max(), color='blue')
plt.plot(k,acumulado[1,:]/acumulado[1,:].max(), color='red')
plt.plot(k,acumulado[2,:]/acumulado[2,:].max(), color='yellow')
plt.plot(k,acumulado[3,:]/acumulado[3,:].max(), color='green')
plt.xlabel('k')
plt.ylabel('p(k)')
plt.title('Distribucin acumulada')
plt.legend(('lamda=1', 'lamda=4', 'lamda=6', 'lamda=10'), loc='lower right')
```

```python
s = np.random.poisson(4, 10000)
plt.hist(s, 14, density=True)
```

### Binomial

```python
n=20
x = np.arange(0,n, 1)
f=np.zeros((len(lamda),len(x)))
acumulado=np.zeros((len(lamda),len(x)))
ac=0
lamda=[.3,.5,.7,.8]
for j in range (len(lamda)):
   for i in range(len(x)):  
      f[j,i]=(math.factorial(n)*(lamda[j]**x[i])*(1-lamda[j])**(n-x[i]))/(math.factorial(x[i])*math.factorial(n-x[i]))
      ac=ac+f[j,i]
      acumulado[j,i]=ac

plt.plot(x,f[0,:], label='F', color='blue')
plt.plot(x,f[1,:], label='F', color='red')
plt.plot(x,f[2,:], label='F', color='yellow')
plt.plot(x,f[3,:], label='F', color='green')

plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Funcin de distribucin bonomial n=100')
plt.legend(('p=0.3', 'p=0.5', 'p=0.7', 'p=0.8'), loc='upper left')
```

```python
s = np.random.binomial(20,0.3, 10000)
plt.hist(s, 14, density=True)
```
