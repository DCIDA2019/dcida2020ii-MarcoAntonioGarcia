```python
import sqlite3
import pandas as pd
db = sqlite3.connect('hubway.db')
def run_query(query):
    return pd.read_sql_query(query,db)
```

### SELECT, LIMIT, WHERE and Etc

Con SELECT podemos especificar columnas por nombre (separadas por comas) o usar el comodín * para devolver cada columna de la tabla.
Además de las columnas que queremos recuperar, también tenemos que decirle a la base de datos de qué tabla obtenerlas. Para hacer esto usamos la palabra clave FROM seguida del nombre de la tabla.

Para seleccionar todas las colmnas usamso SELECT* para seleccionar algunas columnas ponemos el nombre de cabecera

LIMIT simplemente le dice a la base de datos cuántas filas desea que devuelva.

```python
query = 'SELECT * FROM trips LIMIT 5;'  
run_query(query)
```

```python
query = 'SELECT duration, start_date FROM trips LIMIT 5'
run_query(query)
```

Este comando nos permite ordenar la base de datos en una columna determinada.
Para usarlo, simplemente especificamos el nombre de la columna por la que nos gustaría ordenar. De forma predeterminada, ORDER BY ordena en orden ascendente. Si quisiéramos especificar en qué orden se debe ordenar la base de datos, podemos agregar la palabra clave ASC para orden ascendente o DESC para orden descendente.

```python
query = '''SELECT duration FROM trips ORDER BY duration DESC LIMIT 5;
'''
run_query(query)
```

El comando WHERE nos permite usar un operador lógico para especificar qué filas deben devolverse. Por ejemplo, puede utilizar el siguiente comando para devolver todos los viajes realizados con la bicicleta B00400:

```python
query = '''SELECT * FROM trips WHERE (duration > 9990)  ORDER BY duration DESC;'''
run_query(query)
```

```python
query = '''SELECT * FROM trips WHERE (duration >= 9990) AND (gender = "Male") ORDER BY duration DESC; '''
run_query(query)
```

COUNT nos permite trasladar el cálculo a la base de datos y ahorrarnos la molestia de escribir scripts adicionales para contar los resultados. Para usarlo, simplemente incluimos COUNT (column_name) en lugar de (o además de) las columnas que desea SELECCIONAR

```python
query = '''SELECT COUNT(*)FROM trips WHERE sub_type = "Registered";'''
run_query(query)
```

Si queremos que nuestros resultados sean más legibles, podemos usar AS para darle a nuestra salida un alias (o apodo). Volvamos a ejecutar la consulta anterior, pero asignemos al encabezado de nuestra columna un alias de Total de viajes por usuarios registrados:

```python
query = '''SELECT COUNT(*) AS "Total Trips by Registered Users" FROM trips WHERE sub_type = "Registered"; '''
run_query(query)
```

También podemos usar SUM, AVG, MIN y MAX para devolver la suma, promedio, mínimo y máximo de una columna respectivamente. Estos, junto con COUNT, se conocen como funciones agregadas.

```python
query = '''SELECT AVG(duration) AS "Average Duration" FROM trips;'''
run_query(query)
```

GROUP BY separa filas en grupos según el contenido de una columna en particular y nos permite realizar funciones agregadas en cada grupo.

```python
query = '''
SELECT sub_type, AVG(duration) AS "Average Duration" FROM trips GROUP BY sub_type;'''
run_query(query)
```

```python
query = '''SELECT COUNT(*) AS "total de viajes casuales y registrados" FROM trips GROUP BY sub_type;'''
run_query(query)
```

usar operaciones aritméticas directamente dentro de nuestra consulta. SQL nos permite usar +, -, * y / para realizar una operación aritmética en una columna completa a la vez.

```python
query = '''SELECT AVG(duration) FROM trips WHERE (2020 - birth_date) > 30;'''
run_query(query)
```

La base de datos contiene diferentes tablas o podriamos decir que el archivo contiene varias bases de datos, en este caso contiene otra que se llama sations, con la cual es posible trabajar lo antes mencionado solamente llamandola

### JOIN

A continuación, especificamos las tablas que queremos UNIR y le decimos a la base de datos que las conecte en la columna start_station en la tabla de viajes y en la columna de identificación en la tabla de estaciones.
Luego nos adentramos en el meollo de nuestra consulta: AGRUPAMOS POR la ​​columna de la estación en la tabla de estaciones para que nuestro COUNT cuente el número de viajes para cada estación por separado.
Finalmente, podemos ORDENAR POR nuestro RECUENTO y LIMITAR la salida a un número manejable de resultados

```python
query = '''
SELECT stations.station AS "Station", COUNT(*) AS "Count"
FROM trips INNER JOIN stations
ON trips.start_station = stations.id GROUP BY stations.station ORDER BY COUNT(*) DESC
LIMIT 5;
'''
run_query(query)
```

SELECCIONaremos las mismas columnas de salida y UNIREMOS a las tablas de la misma manera, pero esta vez agregaremos una cláusula WHERE para restringir nuestro COUNT a viajes donde start_station es igual que end_station para saber quien hizo un recorrido ida y vuelta.

```python
query = '''SELECT stations.station AS "Station", COUNT(*) AS "Count"
FROM trips INNER JOIN stations
ON trips.start_station = stations.id
WHERE trips.start_station = trips.end_station
GROUP BY stations.station
ORDER BY COUNT(*) DESC
LIMIT 5;
'''
run_query(query)
```
